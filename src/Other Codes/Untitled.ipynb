```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('/Users/eashan22/Desktop/DataMining/FinalProject/Dataset/train_data.csv')
train_labels = pd.read_csv('/Users/eashan22/Desktop/DataMining/FinalProject/Dataset/train_label.csv')
test_data = pd.read_csv('/Users/eashan22/Desktop/DataMining/FinalProject/Dataset/test_data.csv')

total_data = train_data.append(test_data).reset_index(drop=True)
total_data.tail(1)
```

```python
total_data = pd.get_dummies(total_data, dtype=float)
```

```python
total_data.shape
```

```python
total_data = total_data[total_data.columns[60000:67700]]
```

```python
#Split the data into 50% 
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(total_data, test_size = 0.5)
```

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn import metrics
from sklearn.metrics import accuracy_score

%pylab inline 
%matplotlib inline
```

```python
# X_train = train_data

y_train = train_labels
y_train.set_index('id',inplace=True)
y_train.index.rename('id', inplace=True)

# X_test = test_data
y_test = pd.read_csv("/Users/eashan22/Desktop/DataMining/FinalProject/Dataset/submission.csv")
```

```python
# Classifier 4 : Logistic Regression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

model = LogisticRegression()
model.fit(X_train,y_train)
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=3)),
                    ('clf', LogisticRegression())])
pipe_lr.fit(X_train, y_train)
y_pred = model.predict_proba(X_test)

# Classifier 2 : Decision Trees
# from sklearn import tree
# from sklearn.tree import DecisionTreeClassifier

# clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 5000, max_depth=8, min_samples_leaf=2)
# clf_gini.fit(X_train, y_train)
# y_pred = clf_gini.predict_proba(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]

param_grid = [{'clf__C': param_range, }]

gs = GridSearchCV(estimator=pipe_lr, 
                  param_grid=param_grid, 
                  scoring='accuracy', 
                  cv=3,
                  n_jobs=-1)

gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
clf = gs.best_estimator_
clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
gs = GridSearchCV(estimator=pipe_lr,
                  param_grid=param_grid,
                  scoring='accuracy',
                  cv=3)

# Note: Optionally, you could use cv=2 
# in the GridSearchCV above to produce
# the 5 x 2 nested CV that is shown in the figure.

scores = cross_val_score(gs, X_train, y_train, scoring='accuracy', cv=3)
```
