# Is your product listing Attractive?

### Removing the unnecessary components from the table by filtering the colums.

```python
# sLength = len(train_data['id'])
train_data.replace(regex={'<.*?>|&nbsp|\W': ' '}, inplace=True)
train_data = train_data.fillna("missing")
train_data['name'] = train_data['name'].str.lower()
train_data['lvl1'] = train_data['lvl1'].str.lower()
train_data['lvl2'] = train_data['lvl2'].str.lower()
train_data['lvl3'] = train_data['lvl3'].str.lower()
train_data['descrption'] = train_data['descrption'].str.lower()
train_data['type'] = train_data['type'].str.lower()
# train_data['score'] = pd.Series(np.zeros(sLength, dtype=np.int), index=train_data.index)

train_data
```

```python
train_data=train_data.fillna(axis=1,method='ffill')

train_data['type'] = train_data['type'].replace('international', 1)
train_data['type'] = train_data['type'].replace('local', 0)

train_data
```

### Changing the featured data into vector form with word2vec.

```python
train_data.drop(['id'], axis=1, inplace=True)
```

```python
train_data.replace(regex={'<.*?>|&nbsp|\W':' '}, inplace=True)
train_data.head(10)
```

```python
lvl1 = train_data.lvl1.unique()
dic={}
for i,lvl1 in enumerate(lvl1):
    if lvl1 not in dic:
        dic[lvl1]=i
labels=train_data.lvl1.apply(lambda x:dic[x])

train_data.lvl1 = labels

lvl2 = train_data.lvl2.unique()
dic={}
for i,lvl2 in enumerate(lvl2):
    if lvl2 not in dic:
        dic[lvl2]=i+9
labels=train_data.lvl2.apply(lambda x:dic[x])

train_data.lvl2 = labels

lvl3 = train_data.lvl3.unique()
dic={}
for i,lvl3 in enumerate(lvl3):
    if lvl3 not in dic:
        dic[lvl3]=i+66
labels=train_data.lvl3.apply(lambda x:dic[x])

train_data.lvl3 = labels
```

```python
from sklearn.preprocessing import LabelEncoder,LabelBinarizer
from sklearn.preprocessing import OneHotEncoder

le_name = LabelEncoder()
le_descrption = LabelEncoder()
# le_price = LabelEncoder()
# le_type = LabelBinarizer(neg_label=0, pos_label=1, sparse_output=False)

train_data['name'] = le_name.fit_transform(train_data.name.astype(str))
train_data['descrption'] = le_descrption.fit_transform(train_data.descrption.astype(str))
# train_data['price'] = le_price.fit_transform(train_data.price.astype(str))
# train_data['type'] = le_type.fit_transform(train_data.type.astype(str))

# train_data.apply(LabelEncoder().fit_transform)
train_data.head(3)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_boston
import xgboost as xgb

rng = np.random.RandomState(31337)

print("Parallel Parameter optimization")
boston = load_boston()

os.environ["OMP_NUM_THREADS"] = "4"  # or to whatever you want
y_train = boston['target']
X_train = boston['data']
xgb_model = xgb.XGBRegressor()
clf = GridSearchCV(xgb_model, {'max_depth': [2, 4, 6],'n_estimators': [50, 100, 200]}, verbose=1,n_jobs=2)
clf.fit(X_train, y_train)
print(clf.best_score_)
print(clf.best_params_)

y_pred = clf.predict(X_test)
```
